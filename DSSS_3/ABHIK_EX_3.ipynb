```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
a = pd.read_csv("census_income_dataset.csv")
a = a[~a["AGE"].isnull()]
a["AGE"].plot.hist(bins=30,figsize=(8,6),color='thistle',title=("a) Age distribution of respondents"))
plt.xlabel("Age");plt.show();
b=a["RELATIONSHIP"].value_counts()
b.plot(kind='pie',autopct='%1.1f%%',title=("b) Relationship Status"),figsize=(8,6))
plt.ylabel('');plt.show()
a.loc[a["SALARY"] == " <=50K", "SALARY"] = 0
a.loc[a["SALARY"] == " >50K", "SALARY"] = 1
from scipy.stats import pearsonr
corr, p_value = pearsonr(a["EDUCATION-NUM"], a["SALARY"])
a_groups = a.groupby(['EDUCATION'])["SALARY"].sum()
a_groups.plot(kind='bar',figsize=(10,10),ylim=(0,1250),color='teal')
plt.title("c) Educational level and salary")
plt.show()
```
